### **Busca-se dobrar a presen√ßa de doutoras no INSTITUTO DE INFORM√ÅTICA, triplicar na ESCOLA DE ENGENHARIA EL√âTRICA, MEC√ÇNICA E DE COMPUTA√á√ÉO, e quadruplicar na FACULDADE DE MEDICINA, em conson√¢ncia com a busca por inclus√£o e para atender √† crescente demanda do mercado. Quantos doutores adicionais foram contratados?**

```python
import pandas as pd

# Filtrar doutoras no INSTITUTO DE INFORM√ÅTICA
doutoras_informatica = df_ss[(df_ss['formacao'] == 'DOUTORADO') & (df_ss['sexo'] == 'F') & (df_ss['unidade_exercicio'] == 'INSTITUTO DE INFORM√ÅTICA')]

# Filtrar doutoras na ESCOLA DE ENGENHARIA EL√âTRICA, MEC√ÇNICA E DE COMPUTA√á√ÉO
doutoras_engenharia = df_ss[(df_ss['formacao'] == 'DOUTORADO') & (df_ss['sexo'] == 'F') & (df_ss['unidade_exercicio'] == 'ENGENHARIA EL√âTRICA, MEC√ÇNICA E DE COMPUTA√á√ÉO')]

# Filtrar doutoras na FACULDADE DE MEDICINA
doutoras_medicina = df_ss[(df_ss['formacao'] == 'DOUTORADO') & (df_ss['sexo'] == 'F') & (df_ss['unidade_exercicio'] == 'FACULDADE DE MEDICINA')]

# Contar quantidade inicial de doutoras em cada unidade
inicial_informatica = len(doutoras_informatica)
inicial_engenharia = len(doutoras_engenharia)
inicial_medicina = len(doutoras_medicina)

# Duplicar, triplicar e quadruplicar respectivamente a quantidade de doutoras em cada unidade
doutoras_informatica = pd.concat([doutoras_informatica] * 2, ignore_index=True)
doutoras_engenharia = pd.concat([doutoras_engenharia] * 3, ignore_index=True)
doutoras_medicina = pd.concat([doutoras_medicina] * 4, ignore_index=True)

# Contar a quantidade final de doutoras em cada unidade
final_informatica = len(doutoras_informatica)
final_engenharia = len(doutoras_engenharia)
final_medicina = len(doutoras_medicina)

# Calcular o total de doutoras adicionais contratadas
total_adicionais = final_informatica - inicial_informatica + final_engenharia - inicial_engenharia + final_medicina - inicial_medicina

# Imprimir o resultado
print("Total de doutoras adicionais contratadas:", total_adicionais)
```
